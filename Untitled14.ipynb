```python
# Corregir el nombre de la columna y volver a ejecutar el análisis
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("Data.csv")
df_co = pd.read_csv("data_co.csv")

# Unir los datos por Id_Columna
df_merged = df.merge(df_co, on="Id_Columna", how="left")

# Calcular total de preguntas por estudiante y porcentaje de respuestas correctas
df_agg = df_merged.groupby(["Nombre Completo", "Momento"])["Suma de Calificaicón"].agg(
    ["count", "sum"]
).reset_index()
df_agg.rename(columns={"count": "Total_Preguntas", "sum": "Total_Correctas"}, inplace=True)
df_agg["Porcentaje_Correctas"] = df_agg["Total_Correctas"] / df_agg["Total_Preguntas"]

# Clasificación de estudiantes según su porcentaje de respuestas correctas
def clasificar_estudiante(porcentaje):
    if porcentaje >= 0.8:
        return "Alto"
    elif porcentaje >= 0.6:
        return "Medio"
    else:
        return "Bajo"

df_agg["Clasificación"] = df_agg["Porcentaje_Correctas"].apply(clasificar_estudiante)

# Contar estudiantes por clasificación y momento
df_count = df_agg.groupby(["Momento", "Clasificación"]).size().reset_index(name="Cantidad")

# Crear la gráfica
plt.figure(figsize=(10, 6))
sns.barplot(data=df_count, x="Momento", y="Cantidad", hue="Clasificación", palette="viridis")

# Etiquetas y título
plt.xlabel("Momento de Formación")
plt.ylabel("Cantidad de Estudiantes")
plt.title("Clasificación de Estudiantes por Momento de Formación")
plt.legend(title="Clasificación")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Mostrar la gráfica
plt.show()
```
